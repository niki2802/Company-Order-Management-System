# Adding new columns to make the data more use full

# Finding  discount price

```python
df["discount_price"]=df["List Price"]*df["Discount Percent"]/100
```

```python
df.head(5)
```

# Finding the sale price

```python
df["sale_price"]=df["List Price"]-df["discount_price"]
```

```python
df.head(5)
```

# Find the total profit

```python
df["profit"]=df["sale_price"]-df["cost price"]
```
